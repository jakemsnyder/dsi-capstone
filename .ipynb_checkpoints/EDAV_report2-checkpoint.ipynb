```python
test2 = specificdf.groupby(['state']).count().loc[:,'party'].reset_index()
test2.columns =['state', 'num_democrat']
test = pd.concat([test, test2['num_democrat'],test2['num_democrat'],test2['num_democrat']], axis=1)
test.columns = ['state', 'total_num_responses', 'num_dem', 'num_rep', 'num_ind']
test
```

```python
specific_dem = specificdf.loc[specificdf['vote2018']=='Democratic candidate']
specific_rep = specificdf.loc[specificdf['vote2018']=='Republican candidate']
specific_ind = specificdf.loc[specificdf['vote2018']=='Other candidate/not sure']
test2 = specific_dem.groupby(['state']).count().loc[:,'party'].reset_index()
test3 = specific_rep.groupby(['state']).count().loc[:,'party'].reset_index()
test4 = specific_ind.groupby(['state']).count().loc[:,'party'].reset_index()
test['num_dem']=test2['party']
test['num_rep']=test3['party']
test['num_ind']=test4['party']

test
```

```python
# Add proportions 
test['prop_dem']=test['num_dem']/test['total_num_responses']
test['prop_rep']=test['num_rep']/test['total_num_responses']
test['prop_ind']=test['num_ind']/test['total_num_responses']

test
```

```python
import plotly
import plotly.graph_objs as go 

for col in test.columns:
    test[col] = test[col].astype(str)
    
scl = 'RdYlBu'

# scl = [[0.0, 'grey'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
#             [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]

test['text'] = test['state'] + '<br>' +\
    'Number of Responses: '+ test['total_num_responses']+ '<br>' +\
    'Proportion Likely to Vote Democrat: '+ test['prop_dem']+ '<br>' +\
    'Proportion Likely to Vote Republican: '+ test['prop_rep']+ '<br>' +\
    'Proportion Likely to Vote Other: '+ test['prop_ind']+ '<br>'


data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = test['state'],
        locationmode = 'USA-states',
        # z is a numerical value tied to each individual state. In this case, we'll be using 
        # number of people who voted for Trump originally but now disapprove.
        z = test['prop_rep'],
        text = test['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "% Respondents likely to vote Republican")
        ) ]

layout = dict(
        title = '% of Specific Ballot Respondents likely to vote Republican',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
plotly.offline.plot( fig, filename='d3-cloropleth-map.html')
```

```python
# % of folks likely to vote 2020 over time colored by Age
# Step 1: Create month column, group by month, age columns
# Step 2: Calculate Proportion of People who will vote in 2020
# Color by age

df['starttime'] = pd.to_datetime(df['starttime'], unit = 'M')
df['year_month'] = pd.to_datetime(df['starttime'].apply(lambda x: x.strftime('%B-%Y')))

# test4 = df.groupby(['year_month'], as_index=False).count().loc[:, ['year_month','ID']]
# test4.columns = ['year_month', 'num_responses']
# test4['year_month'] = pd.to_datetime(test4['year_month'])
# test4.sort_values(by = 'year_month', ascending = True)
# print(test4)

test5 = df.groupby(['year_month', 'Age', 'Will you vote in the 2020 presidential election?'], as_index=False).count().loc[:, ['year_month', 'Age','Will you vote in the 2020 presidential election?','ID']]
test5.columns = ['year_month', 'Age','will_vote', 'total_num']
test5['will_vote'] = np.nan_to_num(np.array(test5['will_vote']))
print(test5)
```

```python
# Initialize the figure
plt.style.use('seaborn-darkgrid')
 
# create a color palette
palette = plt.get_cmap('Set1')


num=0
for age in test5['Age'].unique():
    num+=1
    #find the right spot on the plot
    plt.subplot(2,3,num)
    
    #plot the lineplot
    plt.plot(test5.loc[test5['Age']==age,'year_month'], test5.loc[test5['Age']==age,'total_num'],color=palette(num), label = age)
    
    #Add title
    plt.title(age, loc = 'left', fontsize = 8,color=palette(num))
    plt.ylabel('count')
    plt.xlabel('year_month')
    plt.xticks(rotation = 'vertical')
    
plt.suptitle('Number of Respondents who will vote in 2020 Election by Age based on Survey Time')

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5,
                    wspace=0.35)
plt.savefig('Count_Age_Willvote2020')
plt.show()
```
